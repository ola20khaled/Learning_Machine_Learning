```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
```

```python
# Reduce the number of features to two using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(iris.data)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, iris.target, test_size=0.25, random_state=42)
```

```python
clf = LogisticRegression(C=1e5)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
# Accuracy
accuracy_score(y_test, y_pred)
```

```python
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred, label="Predicted classes")
plt.xlabel("Sepal length (cm)")
plt.ylabel("Sepal width (cm)")
plt.legend()
plt.show()
```
